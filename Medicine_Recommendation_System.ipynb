```python
### train,test split
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X=df.drop('prognosis',axis=1)
Y=df['prognosis']
```

```python
X
```

```python
Y
```

```python
le=LabelEncoder()
le.fit(Y)
Y=le.transform(Y)
```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=20)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
### Model Training
```

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
##### dictonary of model 
```

```python
models={
"SVC":SVC(kernel='linear'),
 "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42),
    "GradientBoost":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "Knn":KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB":MultinomialNB()
}
```

```python
#printing dictonary
for model_name,model in models.items():
    print(model_name,":",model)
```

```python
for model_name,model in models.items():
    #train model
    model.fit(x_train,y_train)
    #test model
    predictions=model.predict(x_test)
    #accuracy of model
    accuracy=accuracy_score(y_test,predictions)
    #calculate confusion-matrix
    cfmat=confusion_matrix(y_test,predictions)
    #print
    print(f'{model_name} accuracy :{accuracy}')
    
```

```python
#dictonary of symtom and column number
symtoms_dict={}
i=0
for col in X.columns:
    symtoms_dict[col]=i
    i=i+1
```

```python
symtoms_dict.
```

```python
def get_integer_mapping(le):
    '''
    Return a dict mapping labels to their integer values
    from an SKlearn LabelEncoder
    le = a fitted SKlearn LabelEncoder
    '''
    res = {}
    for cl in le.classes_:
        res.update({le.transform([cl])[0]:cl})

    return res
```

```python
#dictonary of label and corresponding encoded value
disease_list = get_integer_mapping(le)
```

```python
disease_list
```

```python
#model prediction function
def get_predicted_value(patient_symptoms):
    input_vector=np.zeros(len(symtoms_dict))
    for item in patient_symptoms:
        input_vector[symtoms_dict[item]]=1
    return disease_list[svc.predict([input_vector])[0]]
```

```python
#helper function
def helper(dis):
    descr=description[description['Disease']==dis]['Description']
    descr=" ".join([w for w in descr])

    pre=precautions[precautions['Disease']==dis][['Precaution_1','Precaution_2','Precaution_3','Precaution_4']]
    pre=[col for col in pre.values]

    med=medication[medication['Disease']==dis]['Medication']
    med=[col for col in med.values]
    
    d8=diet[diet['Disease']==dis]['Diet']
    d8=[col for col in d8.values]

    adv=advice[advice['disease']==dis]['workout']
    return descr,pre,med,d8,adv

```

```python
symtom_inp=input('Enter your symtoms:')
symtom_inp_striped=[s.strip() for s in symtom_inp.split(',')]
symtom_inp_striped=[sym.strip("[]' ") for sym in symtom_inp_striped]
predicted_disease=get_predicted_value(symtom_inp_striped)
descr,pre,med,d8,adv=helper(predicted_disease)

print('*****************************Predicted Disease*****************************')
print(predicted_disease)
print('********************************Description********************************')
print(descr)
print('********************************Precaution********************************')
for pr in pre:
    for p in pr:
        print(p)
print('********************************Medication********************************')
print(med)
print('*******************************Diet*******************************')
print(d8)
print('*******************************Advice*******************************')
for a in adv:
    print(a)
```
